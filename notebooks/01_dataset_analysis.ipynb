```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv("../dataset/prompts_multiclass.csv")

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="category", hue="category", legend=False, palette="viridis", order=df["category"].value_counts().index)

plt.title("Distribución de Clases en el Dataset ", fontsize=15, fontweight="bold")
plt.xlabel("Tipo de Prompt", fontsize=12)
plt.ylabel("Cantidad de Muestras", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

df["length"] = df["prompt"].apply(len)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="length", bins=50, kde=True, color="coral")
plt.title("Distribución de la Longitud de los Prompts", fontsize=15, fontweight="bold")
plt.xlabel("Cantidad de Caracteres", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.show()
```
